### Изменение типов данных и заполнение пропусков

#### Столбец `name`

```python
games_date[games_date['name'].isna()]
```

В столбцах `name` и `genre` пропуски содержаться всего в 2 строках по играм выпущенным на платформе GEN в 1993 году. С одной стоороны можно удалить эти строки, поскольку они не влияют на задачи исследования, но с другой стороны продажи по одной из игр составляют более 2 млн копий, что возможно будет являться значимым при изучении ретроспективы, поэтому логичнее либо оставить пропуски, либо заполнить эти пропуски `unknown`

```python
games_date.loc[games_date['name'].isna(), 'name'] = games_date['name'].fillna('unknown')
```

#### Столбец `year_of_release`

#### Столбец `critic_score`

Большое количество пропусков в данном столбце, а также в столбце с оценками пользователей и рейтингом, может быть связан с тем, что данные оценочные критерии возникли не с самого начала исследуемого периода.

```python
(games_date[(games_date['critic_score'] > 0) & 
           (games_date['year_of_release'] > 0)]
           .sort_values(by='year_of_release').head())
```

```python
games_date[(games_date['critic_score'] > 0) & (games_date['year_of_release'] > 0)].isna().mean()
```

```python
games_date[(games_date['critic_score'].isna()) & (games_date['year_of_release'] > 1996)].head()
```

```python
games_date[games_date['critic_score'] == 0]
```

Действительно оценки критиков появляются на более менее постоянной основе начиная с 1996 года. Изучая пропуски, можно приятик выводу, что пропуски носят несистемный характер и вероятно свзаны с тем, что в каких то регионах или по каким то платформам не проводилась централизованная оценка игр критиками и вообще, данная информация носит скорее дополнительный характер к основной информации. В связи с чем, на мой взгляд, самым корректным будет либо не заполнять эти пропуски, но поскольку в дальнейшем мы будем работать с этими данными, лучше будет заполнить эти пропуски нулями, что необходимо будет учитывать в дальнейшем.

```python
games_date.loc[games_date['critic_score'].isna(), 'critic_score'] = games_date['critic_score'].fillna(0)
games_date['critic_score'] = games_date['critic_score'].astype(float)
```

#### Столбец `user_score`

```python
games_date[games_date['user_score'] == 0]
```

Во многом ситуация с пользовательсткими оценками аналогична ситуации с оценками критиков, единственное глобальное отличие, что помимо пропусков в значениях присутствует аббревиатура 'tbd', которая обозначает To Be Determined (будет определено в дальнейшем). То есть, фактически - это также пропущенные данные, а учитывая, что в ходе исследования необходимо будет производить расчеты со значениями в данном столбце, то считаю возможным преобразовать tbd в нули.

```python
games_date.loc[games_date['user_score'] == 'tbd', 'user_score'] = 0
games_date.loc[games_date['user_score'].isna(), 'user_score'] = games_date['user_score'].fillna(0)
games_date['user_score'] = games_date['user_score'].astype(float)
```

#### Столбец `rating`

Наличие пропусков в столбце с рейтингом игр также во многом связан с тем, что организация ESRB была основана только в 1994 году и для повсеместного присвоения рейтинга ей потребовалось какое то время. Поэтому пропуски в данном столбце оптимальнее заполнить значением unknown

```python
games_date['rating'].sort_values().unique()
```

```python
games_date['rating'] = games_date['rating'].fillna('unknown').astype('category')
```

## Исследовательский анализ данных

### Количество игр по годам

Известно, что часто игры выходят на нескольких платформах одновременно, а поскольку условиями задачи не уточнено необходимо проанализировать общее количество игр или количество уникальных игр вышедших в одном году. Для этого сделаем сводную таблицу с этими количествами по годам.

```python
years_count_games = (games_date[games_date['year_of_release'] > 0]
                    .pivot_table(index='year_of_release', values='name', aggfunc='count')
                    .reset_index())
years_count_game_unique = (games_date[games_date['year_of_release'] > 0]
                        .pivot_table(index='year_of_release', values='name', aggfunc=pd.Series.nunique)
                        .reset_index())


new_col_1 = ['year_of_release', 'count_game']
years_count_games.columns = new_col_1
new_col_2 = ['year_of_release', 'count_game_unique']
years_count_game_unique.columns = new_col_2

years_count_games['year_of_release'] = years_count_games['year_of_release']
years_count_game_unique['year_of_release'] = years_count_game_unique['year_of_release']
years_count_games['count_game_unique'] = years_count_game_unique['count_game_unique']
years_count_games.tail()
```

```python
#Построим гистограмму распределения количества игр по годам

fig, ax = plt.subplots()

ax.bar(years_count_games['year_of_release'], years_count_games['count_game'], 0.9)
ax.bar(years_count_games['year_of_release'], years_count_games['count_game_unique'], 0.8)

ax.set_ylabel('count')
ax.set_xlabel('year')
ax.set_title('count game')

fig.set_figwidth(30)    
fig.set_figheight(15)  

plt.show()
```

```python
years_count_games['count_game'].corr(years_count_games['count_game_unique'])
```

**Вывод** как видно из сводной таблицы и гистограммы:
- начиная с 1991 года игры стали выходить на двух и более платформах, причем рост/снижение количества уникальных игр имеет практически абсолютную корреляцию с общим количеством выпущенных в году игр;
- начиная с 2002 года и по 2009 год наблюдается тенденция к росту количества выпускаемых игр (пик приходится на 2008 и 2009 годы);
- резкое снижение количества выпускаемых игр наблюдается в 2012 году(снижение в сравнении с 2011 годом на 40%);
- гистограмма показывает достаточно ровный тренд на количество игр в период с 2012 по 2016 год, в связи с чем можно сказать, что периоды ранее 2012 года применительно к планированию 2017 года будут нерелевантны.

### Количество платформ по годам

Изучая количество выпущенных игр по годам можно предположить, что подобный пиковый рост и снижение связано с увеличением/снижением количества платформ на которых выходят игры, поэтому мне стало интересно проверить эту гипотезу.

```python
years_count_platform_unique = (games_date[games_date['year_of_release'] > 0]
                               .pivot_table(index='year_of_release', values='platform', aggfunc=pd.Series.nunique)
                               .reset_index()
                              )
```

```python

fig, axes = plt.subplots()
ax = sns.barplot(data=years_count_platform_unique, x='year_of_release', y='platform')
ax.set_ylabel('count platform', fontsize = '20')
ax.set_xlabel('year', fontsize = '20')
ax.set_title('number of platforms by year', fontsize = '20')
fig.set_figwidth(20)    
fig.set_figheight(10)      
```

```python
years_count_platform_unique.tail(20)
```

**Вывод** Исследование показало, что количество игр не имеет особой связи с количеством платформ на которых выходили игры, так в пиковом периоде количества игр, платформ было 7-9, а с 2012 по 2016 год платформ было от 9 до 11, но количество игр снизилось в этот период значительно. Вероятно это говорит о том, что производители стали более тщательно подходить к качеству производимых продуктов (вероятно), снизив при этом их количество. Для проверки этой теории достаточно построить гистограмму распределения суммарной выручки по годам.

### Общие продажи игр по годам

```python
years_sum_sales = (games_date[games_date['year_of_release'] > 0]
                               .pivot_table(index='year_of_release', values='sales_sum', aggfunc=pd.Series.nunique)
                               .reset_index()
                              )
```

```python

fig, axes = plt.subplots()
ax = sns.barplot(data=years_sum_sales, x='year_of_release', y='sales_sum')
ax.set_ylabel('sales', fontsize = '20')
ax.set_xlabel('year', fontsize = '20')
ax.set_title('number of games sold by year', fontsize = '20')
fig.set_figwidth(20)    
fig.set_figheight(10)      
```

**Вывод** В целом нельзя сказать, что снижение количества выпускаемых игр имеет прямую связь с суммарной выручкой от продажи игр, на гистограмме с выручкой отсутсвует аналогичное падение, которое видно на гистограмме с количеством игр в 2012 году, но общий тренд на снижение выручки от продаж игр начиная с 2008 года нельзя не отметить. 

### Общие продажи игр по платформам

#### Сравнение по выручке и количеству релизных лет

Узнаем какие платформы являются лидерами по продажам за весь исследуемый период. Также, сразу проверим гипотезу, что чем больше лет платформа выпускала игры, тем выше ее суммарная выручка в исследуемом периоде.

```python
platform_years_sum_sales = games_date.pivot_table(index='platform', values=['year_of_release', 'sales_sum'], aggfunc={'year_of_release' : pd.Series.nunique, 'sales_sum' : 'sum'}).reset_index()
platform_years_sum_sales = platform_years_sum_sales.sort_values(by='sales_sum', ascending=False).reset_index(drop=True)


fig, axes = plt.subplots()
ax = sns.barplot(data=platform_years_sum_sales, x='platform', y='sales_sum')
ax.set_ylabel('sales', fontsize = '20')
ax.set_xlabel('platform', fontsize = '20')
ax.set_title('sales by platform', fontsize = '20')
fig.set_figwidth(20)    
fig.set_figheight(10)
```

```python
platform_years_sum_sales.head(20)
```

## Проверка гипотез

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковыe

Для проверки гипотезы предлагаются следующие формулировки:
- Нулевая гипотеза - Средние пользовательские рейтинги платформ Xbox One и PC одинаковыe
- Альтернативная гипотеза - Средние пользовательские рейтинги платформ Xbox One и PC отличаются

Предлагается ипользовать для проверки гипотезы критерий alpha 0.05, поскольку это позволит не учитывать возможное влияние выбросов на проводимое исследование

```python
alpha = 0.05 

results = st.ttest_ind(
    actuality_games_date[(actuality_games_date['platform'] == 'XOne') & (actuality_games_date['user_score'] > 0)]['user_score'], 
    actuality_games_date[(actuality_games_date['platform'] == 'PC') & (actuality_games_date['user_score'] > 0)]['user_score'])
          
print('p-значение:', results.pvalue) 

if results.pvalue < alpha: 
    print("Нулевую гипотезу можно отвергнуть")
else:
    print("Нулевую гипотезу нельзя отвергать")
```

**Вывод** по результату проверки можно сделать вывод, что нельзя опровергнуть, что средние пользовательские оценки игр на платформах XOne и PC совпадают.

### Средние пользовательские рейтинги жанров Action  и Sports разные

Для проверки гипотезы предлагаются следующие формулировки:
- Нулевая гипотеза - Средние пользовательские рейтинги жанров Action и Sports различаются
- Альтернативная гипотеза - Средние пользовательские рейтинги жанров Action и Sports совпадают

Предлагается ипользовать для проверки гипотезы критерий alpha 0.05, поскольку это позволит не учитывать возможное влияние выбросов на проводимое исследование

```python
alpha = 0.05

results = st.ttest_ind(
    actuality_games_date[(actuality_games_date['genre'] == 'Action') & (actuality_games_date['user_score'] > 0)]['user_score'], 
    actuality_games_date[(actuality_games_date['genre'] == 'Sports') & (actuality_games_date['user_score'] > 0)]['user_score'])
          
print('p-значение:', results.pvalue) 

if (1 - results.pvalue) < alpha: 
    print("Нулевую гипотезу можно отвергнуть")
else:
    print("Нулевую гипотезу нельзя отвергать")
```

**Вывод** по результату проверки можно сделать вывод, что нельзя опровергнуть, что средние пользовательские оценки игр в жанрах Action и Sports различаются.
